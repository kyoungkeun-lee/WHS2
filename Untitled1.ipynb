```python
# 필요한 라이브러리 설치
!pip install pandas scikit-learn xgboost imbalanced-learn

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
import datetime

# CSV 파일 경로 지정
file_path = '/content/allow_block_logs.csv'  # 실제 경로로 변경하세요

# CSV 파일 로드
waf_logs_df = pd.read_csv(file_path)

# httpSourceId를 숫자로 인코딩
le = LabelEncoder()
waf_logs_df['httpSourceId'] = le.fit_transform(waf_logs_df['httpSourceId'])

# timestamp를 숫자형 특성으로 변환 (예: 유닉스 타임스탬프)
waf_logs_df['timestamp'] = pd.to_datetime(waf_logs_df['timestamp']).astype(int) / 10**9

# responseCodeSent는 이미 숫자형 특성입니다.

# 피처와 레이블 분리
X = waf_logs_df[['httpSourceId', 'timestamp', 'responseCodeSent']]
y = waf_logs_df['action'].apply(lambda x: 1 if x == 'BLOCK' else 0)  # 'BLOCK'을 1로, 'ALLOW'를 0으로 인코딩

# 데이터 균형 확인
print(y.value_counts())

# 'ALLOW' 값이 없는 경우 데이터를 추가하여 균형 맞추기
allow_count = 0
if y.value_counts().get(0, 0) == 0:  # 'ALLOW' 값이 없는 경우
    allow_rows = []
    allow_count = 5000  # 추가할 'ALLOW' 데이터 개수 설정
    for _ in range(allow_count):  # 필요한 개수만큼 'ALLOW' 데이터 추가
        allow_rows.append({
            'httpSourceId': le.fit_transform([f"192.168.{np.random.randint(0, 255)}.{np.random.randint(0, 255)}"])[0],
            'timestamp': datetime.datetime.now().timestamp(),
            'responseCodeSent': 200,
            'action': 'ALLOW'
        })
    allow_df = pd.DataFrame(allow_rows)
    waf_logs_df = pd.concat([waf_logs_df, allow_df], ignore_index=True)

# 추가된 ALLOW 데이터 개수 출력
print(f'Number of ALLOW samples added: {allow_count}')

# 다시 피처와 레이블 분리
X = waf_logs_df[['httpSourceId', 'timestamp', 'responseCodeSent']]
y = waf_logs_df['action'].apply(lambda x: 1 if x == 'BLOCK' else 0)

# 데이터 불균형 해결을 위한 SMOTE 적용
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# 학습 및 테스트 데이터셋 분리
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)

# XGBoost 모델 학습
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 평가
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')

# 데이터셋의 균형 확인
print("Distribution of classes in resampled data:")
print(pd.Series(y_res).value_counts())
```

```python
# 필요한 라이브러리 설치
!pip install pandas scikit-learn xgboost imbalanced-learn

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
import datetime

# CSV 파일 경로 지정
file_path = '/content/allow_block_logs.csv'  # 실제 경로로 변경하세요

# CSV 파일 로드
waf_logs_df = pd.read_csv(file_path)

# httpSourceId를 숫자로 인코딩
le = LabelEncoder()
waf_logs_df['httpSourceId'] = le.fit_transform(waf_logs_df['httpSourceId'])

# timestamp를 숫자형 특성으로 변환 (예: 유닉스 타임스탬프)
waf_logs_df['timestamp'] = pd.to_datetime(waf_logs_df['timestamp']).astype(int) / 10**9

# 추가 특성 로드 (원본 NSL-KDD 데이터셋에서 추가 특성을 사용)
nsl_kdd_file_path = 'KDDTrain+.txt'  # 실제 경로로 변경하세요
columns = ["duration", "protocol_type", "service", "flag", "src_bytes", "dst_bytes", "land", "wrong_fragment",
           "urgent", "hot", "num_failed_logins", "logged_in", "num_compromised", "root_shell", "su_attempted",
           "num_root", "num_file_creations", "num_shells", "num_access_files", "num_outbound_cmds",
           "is_host_login", "is_guest_login", "count", "srv_count", "serror_rate", "srv_serror_rate",
           "rerror_rate", "srv_rerror_rate", "same_srv_rate", "diff_srv_rate", "srv_diff_host_rate",
           "dst_host_count", "dst_host_srv_count", "dst_host_same_srv_rate", "dst_host_diff_srv_rate",
           "dst_host_same_src_port_rate", "dst_host_srv_diff_host_rate", "dst_host_serror_rate",
           "dst_host_srv_serror_rate", "dst_host_rerror_rate", "dst_host_srv_rerror_rate", "label", "difficulty"]

# Use the 'on_bad_lines' parameter to handle problematic lines
nsl_kdd_df = pd.read_csv(nsl_kdd_file_path, names=columns, on_bad_lines='skip')  # 'skip' will skip bad lines

# 필요한 열만 선택
nsl_kdd_df = nsl_kdd_df[["duration", "protocol_type", "service", "flag", "src_bytes", "dst_bytes", "label"]]

# 'ALLOW'와 'BLOCK' 데이터 생성
def create_allow_block_data(row):
    httpSourceId = f"192.168.{np.random.randint(0, 255)}.{np.random.randint(0, 255)}"
    timestamp = datetime.datetime.now().timestamp()
    if row['label'] == 'normal':
        action = 'ALLOW'
        responseCodeSent = 200
    else:
        action = 'BLOCK'
        responseCodeSent = 403
    return [httpSourceId, timestamp, responseCodeSent, action]

# 새로운 데이터프레임 생성
new_data = nsl_kdd_df.apply(create_allow_block_data, axis=1)
new_df = pd.DataFrame(new_data.tolist(), columns=['httpSourceId', 'timestamp', 'responseCodeSent', 'action'])

# 결합을 위해 'label' 열 삭제
nsl_kdd_df = nsl_kdd_df.drop(columns=['label'])

# 기존 waf_logs_df와 nsl_kdd_df 결합
combined_df = pd.concat([waf_logs_df, nsl_kdd_df.reset_index(drop=True)], axis=1)

# 피처와 레이블 분리
X = combined_df.drop(columns=['action'])
y = combined_df['action'].apply(lambda x: 1 if x == 'BLOCK' else 0)

# 데이터 불균형 해결을 위한 SMOTE 적용
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# 학습 및 테스트 데이터셋 분리
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)

# XGBoost 모델 학습
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 평가
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')

# 데이터셋의 균형 확인
print("Distribution of classes in resampled data:")
print(pd.Series(y_res).value_counts())
```
